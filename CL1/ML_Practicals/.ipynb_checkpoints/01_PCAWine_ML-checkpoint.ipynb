```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python

df = pd.read_csv('./datasets/wine.csv')
```

```python
df.head(10)
```

```python

X = df.drop(columns=['Customer_Segment'])  
y = df['Customer_Segment']  
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=X, x='Alcohol', y='Malic_Acid', hue=y,palette='Set2')
plt.title('Scatter Plot (Original Data)')
plt.legend(title='Customer Segment', labels=['Customer Segment 1', 'Customer Segment 2', 'Customer Segment 3'])
plt.show()
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Create a new DataFrame with the first two principal components
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Customer_Segment'] = y
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Customer_Segment', palette='Set1', alpha=0.7)
plt.title('PCA: Wine Dataset')
plt.legend(title='Customer Segment', labels=['Customer Segment 1', 'Customer Segment 2', 'Customer Segment 3'])
plt.show()
```
